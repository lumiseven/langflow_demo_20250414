## 简单的API调用

```python
import requests
import json
url = "http://127.0.0.1:7860/api/v1/run/bab7bc55-6ffb-48a9-af06-c9a65cd11e60"  # The complete API endpoint URL for this flow

# Request payload configuration
payload = {
    "input_value": "hello world!",  # The input value to be processed by the flow
    "output_type": "chat",  # Specifies the expected output format
    "input_type": "chat"  # Specifies the input format
}

# Request headers
headers = {
    "Content-Type": "application/json"
}

try:
    # Send API request
    response = requests.request("POST", url, json=payload, headers=headers)
    response.raise_for_status()  # Raise exception for bad status codes
    print(json.dumps(json.loads(response.text), indent=4, ensure_ascii=False))


    # Print response
    print(response.text)

except requests.exceptions.RequestException as e:
    print(f"Error making API request: {e}")
except ValueError as e:
    print(f"Error parsing response: {e}")
```

## Context 上下文 (Session 会话)

```python
import requests
url = "http://127.0.0.1:7860/api/v1/run/f575fd1b-b849-45e4-b4f0-879fb7c558e7"  # The complete API endpoint URL for this flow

# Request payload configuration
payload = {
    "input_value": "金星呢?",  # The input value to be processed by the flow
    "output_type": "chat",  # Specifies the expected output format
    "input_type": "chat",  # Specifies the input format
    "session_id": "102"  # Optional session ID for tracking
}

# Request headers
headers = {
    "Content-Type": "application/json"
}

try:
    # Send API request
    response = requests.request("POST", url, json=payload, headers=headers)
    response.raise_for_status()  # Raise exception for bad status codes

    # Print response
    print(response.text)

    import json
    print(json.dumps(json.loads(response.text), indent=4, ensure_ascii=False))

except requests.exceptions.RequestException as e:
    print(f"Error making API request: {e}")
except ValueError as e:
    print(f"Error parsing response: {e}")
```
